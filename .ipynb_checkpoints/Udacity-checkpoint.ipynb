<h1 style="color:blue;"> Study 1, can we measure which context is the most efficient? </h1>
<h3> Define a new variable "pages_per_min" which is pages per session_time(min) on a given platform </h3> 

```python
DF['pages_per_min']=DF['total_pages']/(DF['session_duration']/60)
```

```python
pylab.rcParams['figure.figsize'] = (12.0, 6.0) # set size of figures"

quality_cut=(DF['session_duration']>120) & (DF['total_pages']>1) & (DF['pages_per_min']<10)
PPM_info=DF[quality_cut].groupby('context').mean()
PPM_info_std=DF[quality_cut].groupby('context').std()

sorted_index=PPM_info['pages_per_min'].sort_values().index #sorted by average
sorted_means=PPM_info['pages_per_min'].sort_values().values #sorted by average
sorted_std=PPM_info_std['pages_per_min'].sort_values().values #sorted by average

figure()
plt.violinplot([DF[(DF['context']==name) & quality_cut]['pages_per_min'] \
               for name in sorted_index],showmeans=True,showmedians=True,points=1000)

for ii, val in enumerate(sorted_means):
    plt.text(ii+1,val,'{:1.2f}'.format(val)+'$\pm$ '+'{:1.2f}'.format(sorted_std[ii]))

yscale('log')
xticks(range(1,len(PPM_info.index)+1),sorted_index)
ylabel('pages per minute')
```

<h3> Result is plotted in order of efficiency. The standard deviation on the mean is order the size of the mean which indicates furture quality
cuts are needed to remove outliers.</h3>
<h3> No users are actually learning if they are doing 2+ pages per minutes. Also we want to remove those who only viewed one page and left </h3>

<h3> Let's make a basic quality cut. 
<li>total_pages>2</li>
<li>session_duration>120[seconds]</li>
<li> pages_per_min less than 3 </li></h3>

```python
quality_cut=(DF['session_duration']>120) & (DF['total_pages']>2) & (DF['pages_per_min']<3)
PPM_info=DF[quality_cut].groupby('context').mean()
PPM_info_std=DF[quality_cut].groupby('context').std()

sorted_index=PPM_info['pages_per_min'].sort_values().index #sorted by average
sorted_means=PPM_info['pages_per_min'].sort_values().values #sorted by average
sorted_std=PPM_info_std['pages_per_min'].sort_values().values #sorted by average

figure()
plt.violinplot([DF[(DF['context']==name) & quality_cut]['pages_per_min'] \
               for name in sorted_index],showmeans=True,showmedians=True,points=1000)

for ii, val in enumerate(sorted_means):
    plt.text(ii+1,val,'{:1.2f}'.format(val)+'$\pm$ '+'{:1.2f}'.format(sorted_std[ii]))

#yscale('log')
xticks(range(1,len(PPM_info.index)+1),sorted_index)
ylabel('pages per minute')
```

<h3> Conclusion: The most effeicient context is android. On average they go through pages 120% faster than classroom and 60%faster than gae </h3>

<h1 style="color:blue;"> Study 5: Identify dropouts, drop out date and other relavent dropout parameters </h1>
<h3> Define a dropout: group by user and select on no activity for 2 weeks. Further, can we find what lead to a drop out</h3>
<h3>what causes drop out?</h3>
<li>last context used
<li>last course taken
<li>based on trend in efficieny, pages/min ... is it too low, too high ? compare running average
<li>based on time spent (running average of pages and/or session_durration)
<li>days of inactivity
<li>days of inactivity since the start (I notice some students sign up and don't take a course for days)

<h1> step 1) Make a running list of number of days since sign up </h1>
<h3> Group by days_since_sign_up to calculate the relavent stats and join on days_since_sign_up </h3>

```python
DF['start_time_64']=pd.DatetimeIndex(DF['start_time'])
DF['original_timestamp_64']=pd.DatetimeIndex(DF['original_timestamp'])
time_diff=DF['start_time_64']-DF['original_timestamp_64'] #save as nano seconds
#DF['days_since_sign_up']=time_diff.values.astype(float)/1e9/3600/24
DF['days_since_sign_up']=int64(time_diff.values.astype(float)/1e9/3600/24)
```

```python
#view new data frame
DF
```

<h1> Group dataframe by days_since_sign_up </h1>

```python
#first, make a dicionary mapping to tract context for each user. This will be useful when feading into random-forest algorithm
ContMap=dict({('gae',1),('classroom',2),('android',3),('ios',4)})
DF['context_num']=[ContMap[con] for con in DF['context']]
DF_mean_byUser=DF.groupby([DF.index,'days_since_sign_up']).mean()
DF_mean_byUser
```

<h1> We will create a muli-level empty table. This can later be joined on our data frame grouped by [user_id,days_since_sign_up] </h1>
<h3> This will alow us to track the dialy progress (and parameters) of each user on each day </h3>

```python
#define a running list of days for each user to join on:
df_tuples=[]
for user in DF_mean_byUser.index.get_level_values('user_id').values:
    for ii in range(41):
        df_tuples.append((user,ii))
        
index = pd.MultiIndex.from_tuples(df_tuples, names=['user_id', 'days_since_sign_up'])
s = pd.DataFrame(index=index)
s
```

<h3> Merge on both indicies (user_id and days_since_sign_up) and call this the Track dataframe </h3>

```python
Track=s.join(DF_mean_byUser)
```

<h1> Add relavent dropout parameters to the Track dataframe</h1>
<h3>
<li> boolean for inactive: was the user active on day X from sign_up_date? </li>
<li> cumsum_inactive: what is the cumulitive sum of inactiveity since sign up? </li>
<li> days inactive: running tally of number of days since last activity </li>
<li> pages per minute to track efficieny of student in a course </li>
<li> running value for last context used </li>
<li> running value for last course_id taken </li>
<li> cumulative sum_total_pages </li>
<li> running mean of total pages per day </li>
<li> cumulative sum of session duration </li>
<li> running mean of session duration pages per day </li>
</h3>

```python
#define if the user was active or not that day:
Track['inactive']=[isnan(val) for val in Track['total_pages']]
#find cumsum of inactivity
Track['cumsum_inactive']=Track['inactive'].groupby(Track.index.get_level_values('user_id')).cumsum()*Track['inactive']
#black magic to reset the cumsum when the user signs back on :)
sumdiff=Track['cumsum_inactive'].groupby(Track.index.get_level_values('user_id')).diff()
sumdiff[isnan(sumdiff)]=1
sumdiff[sumdiff>1]=1
Track['days_inactive']=sumdiff.groupby(Track.index.get_level_values('user_id')).cumsum()
Track['context_num'].fillna(method='ffill')#forward fill (will ensure we have the last context used)
#we eventually want to do the forward fill for the course ID as well (it will have to be numeric for a machine learning algorythim to handle)
Track=Track.fillna(0) #Also fill NaN with 0.
#Track #(preview dataframe
Track['cumsum_total_pages']=Track['total_pages'].groupby(Track.index.get_level_values('user_id')).cumsum()
Track['cumsum_session_duration']=Track['session_duration'].groupby(Track.index.get_level_values('user_id')).cumsum()
Track['rolling_mean_pages_per_day']=Track['cumsum_total_pages']/Track.index.get_level_values('days_since_sign_up')
Track['rolling_mean_session_duration']=Track['cumsum_session_duration']/Track.index.get_level_values('days_since_sign_up')
Track=Track.fillna(0) #Also fill NaN with 0.
Track
```

<h3> Observation:</h3> 
<h3>Many users are ideal for several days after the initial sign up. The relavent paramter is "cumsum_inactive". If this value ever hits 3 the company should reach out to the user (why did they sign up in the first place if they are not using the service?</h3>
<h3> Another notification should be sent if the "days_inactive" column hits 14 (days since last login) </h3>

<h1> Using the above Track dataframe the dropout users can be defined by the column "days_inactive" </h1>

```python
#all users who dropped out
dropout_user_id=Track.loc[Track['days_inactive'] == 14].index.get_level_values('user_id')
#how many? ... seems like almost all of the given users in this data set
len(dropout_user_id)
```

<h1>Is dropout correlated with course or context? </h1>
<h3> Find the last course and context a particular student took. Eventually use these variables as another metric for the liklihood of dropout </h3>

```python
droupouts=DF[DF.index.isin(set(DF.index).intersection(set(dropout_user_id)))]
#last session time
last_time=droupouts.groupby(droupouts.index).max()['start_time']
a=droupouts.groupby([droupouts.index,'start_time_64']).last()
#a gives us all the corses taken by dropouts. Now let's select on the last corse they took
#last_course_before_drop=a[a.index.get_level_values('start_time_64').isin(last_time)]['course_key']
last_before_drop=a[a.index.get_level_values('start_time_64').isin(last_time)]
```

<h3> count last contexts used by users who dropped out. Exploring if context contributes to dropout liklihood</h3>

```python
#count any of the columns, the user_id is unique in this case
dropout_context_count=last_before_drop.groupby('context').start_time.count()
dropout_context_count
```

<h3> This is of course correlated with the context usage (by unique users) potted above in study 2. To interpret it correclty the count is normalized to number of users of each context </h3>

```python
norm_dropout_contex=dropout_context_count/context_user_count
```

```python
pylab.rcParams['figure.figsize'] = (9.0, 9.0) # set size of figures"
# The slices will be ordered and plotted counter-clockwise.
labels = norm_dropout_contex.index
sizes = norm_dropout_contex.values
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0.01, 0.01, 0, 0.01)  # only "explode" the 2nd slice (i.e. 'Hogs')

patches, texts, autotexts= plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f\%%', pctdistance=0.75, labeldistance=1.05, shadow=True, startangle=90)
plt.title('Contex Before Dropout Normalized to Usage') 
```

<h3> Conclusion: There is an even split in last context used before a student drops out.</h3>
<h3> with ~2000 counts the relative error on the mean is sqrt(2000)/2000 ~= 2%, (except gae where the statistical uncertainty is 0.7%). The differences plotted aboove are not statistically significant. The finding is considtent with uniform lilkihood of dropping out based on context. Yet with more stats this could still be a worthwhile parameter to track in time and use for the dropout likihood</h3>
